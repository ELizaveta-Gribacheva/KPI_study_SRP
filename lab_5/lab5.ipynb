```python


# def harmonic_with_noise(a, amplitude, frequency, phase, noise_mean, noise_covariance, cutoff_frequency, noise_b):
#   #print(phase)
#   w = 2* np.pi * frequency
#   t = np.linspace(0, 10, 1000)
#   y_clean = amplitude * np.sin(w * t + phase)
#   if a == 1:
#     noise = noise_mean + np.sqrt(noise_covariance) * np.random.normal(size=t.shape) 
#     noise_b = noise
#   y_noisy = y_clean + noise_b
#   # noise = np.random.normal(loc=noise_mean, scale=np.sqrt(noise_covariance), size=n)
#   nyquist = 0.5 * (1 / (t[1] - t[0]))  
#   normalized_cutoff = cutoff_frequency / nyquist  # an angular frequency (e.g., rad/s).
#   b, a = iirfilter(N=5, Wn=normalized_cutoff, btype='low', ftype='butter')

#   y_filtered = filtfilt(b, a, y_noisy)

#   return t, y_clean, y_filtered


class CustomSlider(Frame):
    def __init__(self, master=None, width=300, height=50, min_val=0, max_val=10, callback=None):
        Frame.__init__(self, master)
        self.width = width
        self.height = height
        self.min_val = min_val
        self.max_val = max_val
        self.callback = callback

        self.canvas = Canvas(self, width=width, height=height, bg="white", highlightthickness=0)
        self.canvas.pack()
     
        self.track = self.canvas.create_line( 10, height // 2, width - 10, height // 2, width=4, fill="gray")

        slider_img = Image.open("./629031f381125e9e2d18cea9a474353d.png").resize((40, 60), Image.Resampling.LANCZOS)
        self.slider_image = ImageTk.PhotoImage(slider_img)
        self.slider = self.canvas.create_image(10, height // 2, image=self.slider_image, anchor="center")

        self.canvas.tag_bind(self.slider, "<B1-Motion>", self.on_drag)
        self.canvas.tag_bind(self.slider, "<Button-1>", self.on_click)

    def on_drag(self, event):
        x = min(max(event.x, 10), self.width - 10)  
        self.canvas.coords(self.slider, x, self.height // 2)

        if self.callback:
            value = self.get_value(x)
            self.callback(value)

    def on_click(self, event):
        x = min(max(event.x, 10), self.width - 10)
        self.canvas.coords(self.slider, x, self.height // 2)

        if self.callback:
            value = self.get_value(x)
            self.callback(value)
            
    def update_param(self, slider, attr_name):
      new_value = slider.get_value()
      setattr(self, attr_name, new_value)
      self.refresh_figure()

    def set_value(self, value):
      self.value = value
      slider_x = 10 + (value - self.min_val) / (self.max_val - self.min_val) * (self.width - 20)
      
      self.canvas.coords(self.slider, slider_x, self.height // 2)

      if self.callback:
          self.callback(value)


    def get_value(self, x=None):
      if x is None:
          x = self.canvas.coords(self.slider)[0]
      return self.min_val + (x - 10) / (self.width - 20) * (self.max_val - self.min_val)



class SuperShapeFrame(Frame):

  def __init__(self, master = None):

    Frame.__init__(self, master)
    self.grid()

    self.noise_b = None


    self.m = 2
    self.n1 = 0.4
    self.n2 = 1
    self.n3 = 0.1
    self.n4 = 0.3
    self.n5 = 0.4

    self.noise_b = np.zeros(1000)
    self.fig = plt.Figure(figsize=(8, 6), dpi=80)
    self.ax = self.fig.add_subplot(111)
    self.ax.set_title("График гармоники")
    self.ax.grid(True)
    canvas = FigureCanvasTkAgg(self.fig, master=self)
    canvas.get_tk_widget().grid(row=0, column=0, columnspan=4)
    

    self.show_noise = BooleanVar(value=True)
    self.show_filtered = BooleanVar(value=False)
    # self.fig = plt.Figure(figsize=(8, 6), dpi=80)
    # canvas = FigureCanvasTkAgg(self.fig, master=self)
    # canvas.get_tk_widget().grid(row=0, column=0, columnspan=4)

    
    label = Label(self, text = 'Amplitude')
    label.grid(row = 1, column = 1)
    self.amplitude_slider = CustomSlider(self, width=200, height=50, min_val=0, max_val=2, callback=lambda val: self.update_param(self.amplitude_slider, "m"))

#elf.amplitude_slider = Scale(self, from_ = 0, to = 2, resolution=0.001, length=200, orient = HORIZONTAL, command = lambda i : self.update_m())

    self.amplitude_slider.grid(row = 1, column = 2)

    label = Label(self, text = 'Frequency')
    label.grid(row = 2, column = 1)
    self.n1_slider = Scale(self, from_ = 0, to = 1, resolution=0.001, length=200, orient = HORIZONTAL, command = lambda i : self.update_n1())
    self.n1_slider.grid(row = 2, column = 2)

    label = Label(self, text = 'Phase')
    label.grid(row = 3, column = 1)
    self.n2_slider = Scale(self, from_ = 0, to = 1, resolution=0.001, length=200, orient = HORIZONTAL, command = lambda i : self.update_n2())
    self.n2_slider.grid(row = 3, column = 2)

    label = Label(self, text = 'Noise Mean')
    label.grid(row = 4, column = 1)
    self.n3_slider = Scale(self, from_ = 0, to = 0.2, resolution=0.001,length=200, orient = HORIZONTAL, command = lambda i : self.update_n3())
    self.n3_slider.grid(row = 4, column = 2)

    label = Label(self, text = 'Noise Covariance')
    label.grid(row = 5, column = 1)
    self.n4_slider = Scale(self, from_ = 0, to = 1, resolution=0.001, length=200, orient = HORIZONTAL, command = lambda i : self.update_n4())
    self.n4_slider.grid(row = 5, column = 2)

    label = Label(self, text = 'Cutoff Frequency')
    label.grid(row = 6, column = 1)
    self.n5_slider = Scale(self, from_ = 0, to = 10, resolution=0.001, length=200, orient = HORIZONTAL, command = lambda i : self.update_n5())
    self.n5_slider.grid(row = 6, column = 2)

    self.noise_checkbox = Checkbutton(self, text="Показывать шум", variable=self.show_noise, command=self.refresh_figure)
    self.noise_checkbox.grid(row=7, column=3, columnspan=3)

    self.noise_checkbox = Checkbutton(self, text="Відфільрована", variable=self.show_filtered, command=self.refresh_figure)
    self.noise_checkbox.grid(row=8, column=3, columnspan=3)

    img = Image.open(r"./ic_launcher.png")
    self.photo = ImageTk.PhotoImage(img)

    label = Label(self, text = 'Reset')
    label.grid(row = 7, column = 1) 
    self.button = Button(self, text="Reset", command=self.f, image=self.photo)
    self.button.grid(row=7, column=2)
    

    self.draw_initial_figure()

  def harmonic_with_noise(self, a, amplitude, frequency, phase, noise_mean, noise_covariance, cutoff_frequency):
      #print(phase)
      w = 2* np.pi * frequency
      t = np.linspace(0, 10, 1000)
      y_clean = amplitude * np.sin(w * t + phase)
      if a == 1:
        noise = noise_mean + np.sqrt(noise_covariance) * np.random.normal(size=t.shape) 
        self.noise_b = noise
      y_noisy = y_clean + self.noise_b
      # noise = np.random.normal(loc=noise_mean, scale=np.sqrt(noise_covariance), size=n)
      nyquist = 0.5 * (1 / (t[1] - t[0]))  
      normalized_cutoff = cutoff_frequency / nyquist  # an angular frequency (e.g., rad/s).
      b, a = iirfilter(N=5, Wn=normalized_cutoff, btype='low', ftype='butter')

      y_filtered = filtfilt(b, a, y_noisy)

      return t, y_clean, y_noisy, y_filtered
  


  # def f(self):
  #   self.button['bg'] = '#A4C639'          
  #   self.button['activebackground'] = '#E32636'
    
  def f(self):
        print("Reset")
        # msg=messagebox.showinfo( "Plot reset")
        
        self.button.update()
        self.m = 0.97
        self.n1 = 0.267
        self.n2 = 0
        self.n3 = 0.108
        self.n4 = 0.101
        self.n5 = 5.0

        self.amplitude_slider.set_value(self.m)
        self.n1_slider.set(self.n1)
        self.n2_slider.set(self.n2)
        self.n3_slider.set(self.n3)
        self.n4_slider.set(self.n4)
        self.n5_slider.set(self.n5)
        self.refresh_figure(0)

  def update_param(self, slider, attr_name):
    setattr(self, attr_name, slider.get_value())
    self.refresh_figure(0)

  # def update_m(self, value):
  #   self.m = value
  #   # self.m = self.amplitude_slider.get()
  #   self.refresh_figure()

  def update_n1(self):
    print(self.n1_slider.get())
    self.n1 = self.n1_slider.get()
    print(self.n1)
    self.refresh_figure(0)

  def update_n2(self):
    self.n2 = (self.n2_slider.get())
    self.refresh_figure(0)

  def update_n3(self):
    self.n3 = self.n3_slider.get()
    self.refresh_figure(1)

  def update_n4(self):
    self.n4 = self.n4_slider.get()
    self.refresh_figure(1)

  def update_n5(self):
    self.n5 = self.n5_slider.get()
    self.refresh_figure(1)

  def refresh_figure(self, a):

    t, y_clean,y_noisy, y_f  = self.harmonic_with_noise(a, self.m, self.n1, self.n2, self.n3, self.n4, self.n5)

    
    self.ax.cla()
    # self.lines_clean, = self.ax.plot(t, y_clean, "b-", label="Чистая гармоника")

    if self.show_filtered.get():
        self.lines_clean, = self.ax.plot(t, y_f, "m-", label="Filtered data")
        self.ax.plot(t, y_clean, "b--", label="Чистая гармоника")
    else:
       self.lines_clean, = self.ax.plot(t, y_clean, "b--", label="Чистая гармоника")
    
    if self.show_noise.get():
        self.lines_noisy, = self.ax.plot(t, y_noisy, "orange", alpha=0.6, label="Шум")
    else:
        self.lines_noisy, = self.ax.plot([], [], "orange", alpha=0.6, label="Шум")  

    self.ax.set_xlim(min(t) - 1, max(t) + 1)
    self.ax.set_ylim(min(y_clean) - 1, max(y_clean) + 1)

    if hasattr(self, 'icon_img'):
        self.icon_img.remove()
        self.add_icon_to_plot(t[0], y_clean[0])
        
    self.ax.legend()
    self.ax.grid(True)
    
    self.fig.canvas.draw()


  def draw_initial_figure(self):
    t, y_clean, y_noisy, y_f = self.harmonic_with_noise(1,self.m, self.n1, self.n2, self.n3, self.n4, self.n5)
    
    if not hasattr(self, 'ax'):
        self.ax = self.fig.add_subplot(111)
    else:
        self.ax.cla()

    self.lines_clean, = self.ax.plot(t, y_clean, "b-", label="Чистая гармоника")
    self.lines_noisy, = self.ax.plot(t, y_noisy, "orange", alpha=0.6, label="Шум")

    self.ax.set_xlim(min(t) - 1, max(t) + 1)
    self.ax.set_ylim(min(y_clean) - 1, max(y_clean) + 1) 

    self.add_icon_to_plot(t[0], y_clean[0])
    
    self.ax.legend()
    self.fig.canvas.draw()


  def add_icon_to_plot(self, x, y):
        img = Image.open(r"./52a6e0cd04260b88240ff565e8d7e855.png")
        img = img.resize((30, 40), Image.Resampling.LANCZOS)  
        img_array = np.array(img)

        wid = 0.5  
        height = 0.5  
        x_extent = [x - wid / 2, x + wid / 2]
        y_extent = [y - height / 2, y + height / 2]

        self.icon_img = self.ax.imshow( img_array,extent=x_extent + y_extent,aspect='auto',zorder=10)
        self.fig.canvas.draw()


  
app = SuperShapeFrame()
app.master.title('harmonic function')
app.mainloop()
```

An IIR filter is described by a so-called difference equation, which defines
how an incoming signal is processed over time. A new output $y[n]$ is
obtained by mathematically combining past filter outputs with past inputs (recursion)1.
A difference equation looks something like this:
$$
a_0 y[n] = b_0 x[n] + b_1 x[n-1] + b_2 x[n-2] - a_1 y[n-1] - a_2 y[n-2]
$$
